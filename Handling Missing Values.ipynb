```python
#importing dependencies
import pandas as pd
import numpy as np
```

```python
#read data to a dataframe
nfl_data = pd.read_csv("Datasets/NFL Play by Play 2009-2017 (v4).csv")
#setting seed for reproducibility
np.random.seed(0)
```

```python
#taking a look at the dataset
nfl_data.head()
```

```python
'''
We observe that some don't have values they are indicated by NaN or None
'''
# counting number of missing data per column

missing_values_count = nfl_data.isnull().sum()
print(missing_values_count)
```

```python
#looking at first 10 columns and their count of missing values
print(missing_values_count[0:10])
```

```python
# computing total number of cells and cells with miisng values
total_cells = np.product(nfl_data.shape)
missing_cells = missing_values_count.sum()
missing_percent = (missing_cells/total_cells) * 100
print(missing_percent)
```

```python
'''
The data can be missing if it wasn't recorded or if it doesn't exist.
If it doesn't exist then it makes sense to keep it as NaN,
Whereas if it exists but wssn't recorded then they should be replaced by a suitable value
'''
```

```python
'''
In some cases trying to figure out the reason why data wass missing can be costly or not required, that time we can use dropna function to remove columns/rws with missing values
It is not highly recomended though
'''
nfl_data.dropna()
```

```python
'''
Since every row had a missing value dropna deleted all the rows.
So let's consider columns and drop columns instead of rows
'''
columns_with_na_dropped = nfl_data.dropna(axis=1) # axis=1 tells to consider clumns instead of rows
columns_with_na_dropped.head()
```

```python
'''
Now let's see how many columns were dropped
'''
columns_dropped = columns_with_na_dropped.shape[1]
total_columns = nfl_data.shape[1]
print("Columns in original dataset: %d" % total_columns)
print("Columns dropped: %d" %columns_dropped)
```

```python
#Filling in missing values automatically

#getting a small subset of the data
subset_nfl_data = nfl_data.loc[:,'EPA':'Season'].head()
```

```python
subset_nfl_data
```

```python
# replacing NaNs with zero
subset_nfl_data.fillna(0)
```

```python
#replacing NaNs with the next occuring value in the column
subset_nfl_data.fillna(method='bfill',axis=0).fillna(0)
#bfill denotes to replace with next occuring item
#another fillna(0) is applied to tackle the corner case of NaN in the last element of the column
```
